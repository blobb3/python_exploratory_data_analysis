Diese Kreuztabelle zeigt eine Verteilung der Kraftstoffarten (Spalte „fuel_type“) in Verbindung mit verschiedenen Getriebearten (Spalte „transmission“) in einem Datensatz über Autos. Die Werte in der Tabelle geben die Anzahl der Fahrzeuge an, die jeweils eine bestimmte Kombination von Kraftstoff- und Getriebetyp haben. 

### Interpretation der Hauptpunkte:

- **Benzin**:
  - Die meisten Benzinfahrzeuge haben ein **manuelles Schaltgetriebe** (1070) und **automatisierte Schaltgetriebe** (713).
  - Es gibt auch eine beträchtliche Anzahl von Benzinfahrzeugen mit **Automatik** (244) und **Automat stufenlos** (487).

- **Diesel**:
  - Die meisten Dieselfahrzeuge haben ebenfalls ein **manuelles Schaltgetriebe** (306), gefolgt von **automatisierten Schaltgetrieben** (198).
  - Ein relativ kleiner Teil der Dieselfahrzeuge hat **Automatik** (64) oder **Automat stufenlos** (564).

- **Elektro**:
  - Elektrofahrzeuge kommen hauptsächlich mit **Automatikgetriebe** (58), was typisch für Elektrofahrzeuge ist.

- **Alternative Kraftstoffe (z.B., Erdgas, Flüssiggas, Hybrid)**:
  - Es gibt nur wenige Fahrzeuge mit alternativen Kraftstoffen wie **Erdgas (CNG) / Benzin** und **Flüssiggas (LPG) / Benzin** und auch **Hybrid Benzin/Elektro**.
  - Diese haben typischerweise entweder **Schaltgetriebe** oder ein **automatisiertes Schaltgetriebe**.

### Allgemeine Schlussfolgerungen:
- **Manuelles Schaltgetriebe** ist bei Benzin- und Dieselfahrzeugen am häufigsten vertreten.
- **Automatikgetriebe** ist verbreitet bei Elektrofahrzeugen und Benzinfahrzeugen.
- **Alternative Kraftstoffe** sind in der Stichprobe weniger häufig und weisen eine geringere Vielfalt an Getriebearten auf.

Diese Tabelle kann genutzt werden, um die Präferenzen und Verteilungen der Getriebetypen in Verbindung mit den Kraftstoffarten zu analysieren.

### Pivot tables

Dieser Code erstellt eine Pivot-Tabelle, die den Durchschnitt (mean) und die Anzahl (count) der Variablen **hp** (Pferdestärke), **mileage** (Kilometerstand) und **price** (Preis) für verschiedene Kraftstoffarten (**fuel_type**) berechnet. Die Tabelle fasst die Mittelwerte und Anzahl der Autos in jeder Kraftstoffkategorie zusammen, wodurch eine übersichtliche Darstellung der typischen Leistungswerte, Kilometerstände und Preise je Kraftstoffart entsteht.

### Correlation matrix

```python
corr = df[['mileage', 
          'price', 
          'hp', 
]].corr()
corr
```

Die Korrelationsmatrix zeigt den Zusammenhang zwischen den Variablen **mileage** (Kilometerstand), **price** (Preis) und **hp** (Pferdestärke):

1. **Mileage vs. Price**: Die Korrelation ist -0.44, was auf eine moderate negative Beziehung hinweist. Das bedeutet, dass Autos mit höherem Kilometerstand tendenziell günstiger sind.
   
2. **Price vs. HP**: Die Korrelation beträgt 0.61, was auf eine moderate positive Beziehung hinweist. Autos mit höherer PS-Zahl haben also tendenziell höhere Preise.
   
3. **Mileage vs. HP**: Die Korrelation ist -0.13, was auf eine schwache negative Beziehung hinweist. Es besteht kaum ein Zusammenhang zwischen Kilometerstand und PS-Zahl.

Insgesamt zeigt die Matrix, dass **Preis** und **PS** am stärksten miteinander korrelieren, während der **Kilometerstand** eine schwächere Beziehung zu beiden anderen Variablen hat.

### Covariance matrix

```python
cov = df[['mileage', 
          'price', 
          'hp', 
]].cov()
cov
```

Die Kovarianzmatrix zeigt die Streuung zwischen den Variablen **mileage** (Kilometerstand), **price** (Preis) und **hp** (Pferdestärke). Hier ist eine kurze Interpretation:

1. **Diagonalwerte** (z. B. **Mileage** zu **Mileage**): Diese Werte zeigen die Varianz jeder Variablen. Zum Beispiel hat **mileage** eine hohe Varianz (3.73e+09), was bedeutet, dass der Kilometerstand in der Datenverteilung stark streut.

2. **Off-Diagonalwerte**:
   - **Mileage und Price**: Die Kovarianz beträgt -7.12e+08, was auf eine negative Beziehung hindeutet, d. h., Autos mit höherem Kilometerstand tendieren zu niedrigeren Preisen.
   - **Price und HP**: Die Kovarianz beträgt 1.96e+06, was eine positive Beziehung andeutet, d. h., Autos mit mehr PS sind tendenziell teurer.
   - **Mileage und HP**: Die Kovarianz ist -9.24e+05, was auf eine sehr schwache negative Beziehung hinweist.

Zusammengefasst zeigt die Kovarianzmatrix ähnlich wie die Korrelationsmatrix, dass **Preis** und **PS** tendenziell positiv zusammenhängen, während **Kilometerstand** und **Preis** negativ zusammenhängen. Die Werte der Kovarianz sind jedoch skalenabhängig und schwerer zu interpretieren als Korrelationen.

## Univariate graphical exploratory data analysis (EDA)

### Line chart (matplotlib)

```python
# Generate time series data for mean car prices over the last 20 years
x = []
y = []

# Start from 25 years ago with an initial price
start_date_trend = datetime(2000, 1, 1)
initial_price = 100

for i in range(25):
    date = start_date_trend + timedelta(days=365 * i)
    price_increase_factor = 1 + random.uniform(0.01, 0.02)
    initial_price *= price_increase_factor
    x.append(date)
    y.append(round(initial_price, 2))

# Create a DataFrame to display the data
rental_prices_trend_df = pd.DataFrame({
    "Year": x,
    "Price (USD)": y
})


# Create figure
fig = plt.figure(figsize=(7,4))
plt.plot(x, y, color="green")
plt.title('Development of mean car prices', fontsize=12)
plt.xlabel('year', fontsize=10)
plt.ylabel('price index (1990 = 100%)', fontsize=10)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)

plt.show()
```

**Zusammenfassung**: Der Density Plot bestätigt, dass die Mehrheit der Autos preisgünstig ist, während teurere Fahrzeuge in geringer Anzahl vertreten sind. Die Glättung der Dichtekurve bietet eine visuelle Darstellung der Preisverteilung und zeigt klar, dass der Preisschwerpunkt im unteren Bereich liegt.

### Quantile-Quantile (QQ-) plot

```python
# Variable (standardized)
x = 'mileage'
df_qq = df
df_qq['var'] = (df[x]-df[x].mean()) / df[x].std()
print(df_qq.sort_values('var')[['mileage', 'var']])
    
# Plot
sm.qqplot(df_qq['var'], line ='45')
plt.show
```

Der Quantile-Quantile-Plot (QQ-Plot) hier zeigt, wie die Verteilung der `mileage`-Daten (in standardisierter Form) im Vergleich zu einer theoretischen Normalverteilung aussieht.

### Interpretation des QQ-Plots:
- Die `mileage`-Datenpunkte werden auf der x-Achse gegen die quantilenormalisierten Daten auf der y-Achse aufgetragen.
- Die rote Linie (45°-Linie) stellt eine perfekt normalverteilte Linie dar.
- Wenn die Daten der Normalverteilung entsprechen, sollten die Punkte weitgehend entlang dieser Linie verlaufen.

#### Beobachtungen:
- Im unteren und mittleren Bereich (Quantilen um 0) verlaufen die Datenpunkte recht nah an der Linie, was darauf hinweist, dass dieser Bereich der `mileage`-Verteilung annähernd normalverteilt ist.
- Ab den höheren Quantilen (ungefähr ab 2) entfernen sich die Datenpunkte deutlich von der Linie nach oben. Dies zeigt, dass die `mileage`-Verteilung am oberen Ende rechtsschief ist und extreme Werte (Outliers) in den hohen `mileage`-Werten aufweist.

Zusammengefasst zeigt dieser QQ-Plot, dass die `mileage`-Verteilung nicht perfekt normalverteilt ist und eine rechtsschiefe Tendenz aufweist, insbesondere durch hohe `mileage`-Werte (die Outliers sein könnten).

Diese Heatmap visualisiert die Korrelationen zwischen den Variablen *mileage* (Kilometerstand), *hp* (PS) und *price* (Preis) des Datensatzes.

**Interpretation der Korrelationen:**
1. **mileage und price**: Der Korrelationskoeffizient beträgt -0,44. Das bedeutet, dass ein negativer Zusammenhang besteht – je höher der Kilometerstand eines Fahrzeugs ist, desto niedriger ist im Allgemeinen der Preis. Dies ist ein gängiges Muster, da Fahrzeuge mit höherem Kilometerstand in der Regel weniger wert sind.
   
2. **hp und price**: Hier ist die Korrelation mit 0,61 positiv und relativ stark. Das bedeutet, dass Autos mit mehr PS tendenziell teurer sind. Dies ist ebenfalls intuitiv, da leistungsstärkere Autos oft teurer sind.

3. **mileage und hp**: Die Korrelation zwischen Kilometerstand und PS ist mit -0,13 schwach und negativ. Dies weist darauf hin, dass es keine starke Beziehung zwischen diesen beiden Variablen gibt.

**Zusammengefasst**:
- Die Preis-Kilometerstand- und Preis-PS-Beziehungen sind am aussagekräftigsten.
- Die Farben der Heatmap (Braun und Türkis) unterstreichen die Richtung und Stärke der Korrelationen (dunkelbraun für negative und türkis für positive Korrelationen).

### Bubble plot (seaborn)

```python
# Calculate price per hp
df['price_per_hp'] = df['price'] / df['hp']

# Subset of df
df_sub = df.loc[(df['price'] >= 25000) & (df['price'] <= 100000)]

plt.figure(figsize=(7,4))
plt.ticklabel_format(style='plain')
cmap = sns.cubehelix_palette(dark=.3, light=3, as_cmap=True)

ax = sns.scatterplot(x="mileage", 
                     y="price",
                     size="price_per_hp", # determines bubble size
                     hue="mileage",  # determines color
                     palette="Set2",
                     data=df_sub)

# Set title and axes
ax.set_title('Price vs mileage', fontsize = 12)
ax.set_xlabel('mileage', fontsize = 10)
ax.set_ylabel('price', fontsize = 10)
ax.legend([],[], frameon=False) # skip legend
```

Dieser Bubble-Plot zeigt die Beziehung zwischen dem Kilometerstand (x-Achse) und dem Preis (y-Achse) der Autos. Die Datenpunkte (Blasen) repräsentieren Autos, deren Preise zwischen 25.000 und 100.000 liegen.

**Interpretation:**

1. **Blasengröße**: Die Größe der Blasen wird durch den "Preis pro PS" (price per hp) bestimmt. Größere Blasen zeigen Autos mit einem höheren Preis pro PS, was auf teurere und möglicherweise leistungsfähigere Fahrzeuge hindeutet.

2. **Farbgebung**: Die Farbe der Blasen spiegelt den Kilometerstand wider, wobei der Farbübergang verschiedene Werte repräsentiert. Dies bietet eine zusätzliche visuelle Dimension, um zu erkennen, wie sich der Kilometerstand auf den Preis auswirken könnte.

3. **Verteilung**: Die Daten zeigen, dass Autos mit niedrigerem Kilometerstand tendenziell teurer sind. Autos mit höherem Kilometerstand befinden sich hauptsächlich im unteren Preisbereich.

Dieser Plot hilft zu erkennen, wie sich Kilometerstand und Preis pro PS auf den Verkaufspreis der Autos auswirken. Fahrzeuge mit geringem Kilometerstand haben tendenziell einen höheren Preis, während der Preis bei höherem Kilometerstand sinkt.
